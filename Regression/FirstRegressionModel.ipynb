## My First Regression Model

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

```python
# Load the Boston data set
dataset = datasets.load_boston()
```

```python
dataset.keys()
```

```python
print(dataset['DESCR'])
```

```python
# Prepare the data

# The target is "MEDV: Median value of owner-occupied homes in $1000's"
y = dataset['target']

# Use only "RM: average number of rooms per dwelling" as a feature
x = dataset['data'][:, [5]] # take every row but only 5th col. if want more than 1 feature add more col number to list, eg. [1,5]
```

```python
# Visualize the data
plt.scatter(x, y)
plt.xlabel('RM')
plt.ylabel('Price')

# Visualize to see the trend of the data, to then use to decide which model to use
```

```python
# Init a linear model
model = linear_model.LinearRegression()

# Train the model
model.fit(x, y) # calling a learning algorithm

# Training time depend on data's size and the size of the model (model parameter)
```

```python
print('Price =', model.coef_, '* RM +', model.intercept_)
```

```python
y_pred = model.coef_ * x + model.intercept_
```

```python
plt.scatter(x, y, color='black')
plt.plot(x, y_pred, color='blue', linewidth=3)
plt.xlabel('RM')
plt.ylabel('Price')
```

```python
# Make a prediction for a dwelling with 7 rooms
print(model.coef_ * 7 + model.intercept_)
```

```python
print(model.predict([[7]]))
# input is 2d array because can predict multiple samples at once and for multiple features
```

```python
# Make predictions for all samples
print(model.predict(x))
```
