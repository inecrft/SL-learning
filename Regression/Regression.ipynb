## Regression Model

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

```python
# Load the California housing dataset
dataset = datasets.fetch_california_housing()
```

```python
dataset.keys()
```

```python
print(dataset['DESCR'])
```

```python
# Prepare the data

# The target is "median house value for California districts in $100,000"
y = dataset['target']

# Use only "MedInc: median income in block group" as a feature
x = dataset['data'][:, [0]] # take every row but only 0th col. if want more than 1 feature add more col number to list, eg. [1,5]
```

```python
# Visualize the data
plt.scatter(x, y)
plt.xlabel('MedInc')
plt.ylabel('Price')

# Visualize to see the trend of the data, to then use to decide which model to use
```

```python
# Init a linear model
model = linear_model.LinearRegression()

# Train the model
model.fit(x, y) # calling a learning algorithm

# Training time depend on data's size and the size of the model (model parameter)
```

```python
print('Price =', model.coef_, '* MedInc +', model.intercept_)
```

```python
y_pred = model.coef_ * x + model.intercept_
```

```python
plt.scatter(x, y, color='black')
plt.plot(x, y_pred, color='blue', linewidth=3)
plt.xlabel('MedInc')
plt.ylabel('Price')
```

```python
# Make a prediction for a housing with income of 7
print(model.coef_ * 7 + model.intercept_)
```

```python
print(model.predict([[7]]))
# input is 2d array because can predict multiple samples at once and for multiple features
```

```python
# Make predictions for all samples
print(model.predict(x))
```
